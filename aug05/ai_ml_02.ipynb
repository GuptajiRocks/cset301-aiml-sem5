```python
import pandas as pd
import matplotlib.pyplot as plt
import cv2
testi = "test.jpg"
```

```python
def dispFunc(img):
  plt.imshow(img)
  plt.axis("off")

def dispIg(img):
  plt.imshow(img, cmap="gray")
  plt.axis("off")
```

```python
img = cv2.imread(testi)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis("off")
```

```python
img_resize = cv2.resize(img, (300,300))
plt.imshow(img_resize)
plt.axis("off")
```

```python
gray_image = cv2.cvtColor(img_resize, cv2.COLOR_RGB2GRAY)
dispIg(gray_image)
```

```python
thresh = 100
max_val = 255
_, bin_img = cv2.threshold(gray_image, thresh, max_val, cv2.THRESH_BINARY)
plt.subplot(1, 2, 1)
dispIg(bin_img)
plt.subplot(1,2, 2)
dispFunc(bin_img)
```

```python
normalized = gray_image / 255.0
print("Normalized image shape:", normalized.shape)
print("Min pixel value:", normalized.min())
print("Max pixel value:", normalized.max())
print("Min pixel value:", gray_image.min())
print("Max pixel value:", gray_image.max())
```

```python
fh = cv2.flip(gray_image, 1)
fp = cv2.flip(gray_image, 0)
r9 = cv2.rotate(gray_image, cv2.ROTATE_90_CLOCKWISE)
r180 = cv2.rotate(gray_image, cv2.ROTATE_180)

plt.subplot(2,2,1)
dispFunc(fh)
plt.subplot(2,2,2)
dispFunc(fp)
plt.subplot(2,2,3)
dispFunc(r9)
plt.subplot(2,2,4)
dispFunc(r180)

plt.show()
```

```python
gauss_bl = cv2.GaussianBlur(gray_image, (3,3), 0)
dispIg(gauss_bl)
```
